```python
# отключим предупреждения Anaconda
import warnings
warnings.simplefilter('ignore')

# будем отображать графики прямо в jupyter'e
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#графики в svg выглядят более четкими
%config InlineBackend.figure_format = 'svg' 

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
import numpy as np
import pandas as pd
```

# Создание датасета

```python

import random
random.seed(42)
investors = []
for i in range(1, 101):
    investors.append('investor' + str(i))
investor_ids = [i + 1 for i in range(len(investors))]

businesses = []
for i in range(1, 101):
    businesses.append('business' + str(i))

PV = [random.randint(50000, 500000) for _ in range(len(investors))]

df = pd.DataFrame()

deposit_rate = 0.061
for i in range(100):
        a = random.uniform(-0.9, 0.5)
        income_year1 = PV[i] + PV[i] * a
        b = random.uniform(-0.2, 0.7)
        income_year2 = income_year1 + income_year1 * b
        c = random.uniform(0.05, 0.3)
        income_year3 = income_year2 + income_year2 * c
        d = random.uniform(0.1, 0.3)
        income_year4 = income_year3 + income_year3 * d
        Total_profit = round((income_year4 - PV[i])/(PV[i]) * 100, 1)


        deposit_rate1 = random.uniform(0.055, 0.15)
        deposit_rate2 = random.uniform(0.055, 0.15)
        deposit_rate3 = random.uniform(0.055, 0.15)
        deposit_rate4 = random.uniform(0.055, 0.15)
        FV1_deposit = PV[i] + PV[i] * deposit_rate1
        FV2_deposit = FV1_deposit + FV1_deposit * deposit_rate2
        FV3_deposit = FV2_deposit + FV2_deposit * deposit_rate3
        FV4_deposit = FV3_deposit + FV3_deposit * deposit_rate4
        Total_profit_deposit = round((FV4_deposit - PV[i])/(PV[i]) * 100,1)
        new_row = pd.DataFrame({
            'Investor': [investors[i]],
            'ID of investor': [investor_ids[i]],
            'Business': [businesses[i]],
            'Present value': PV[i],
            'Deposit_rate1 %' : round(deposit_rate1 * 100, 1),
            'Rate_business1 %': round(a * 100, 1),
            'FV1_Business': int(income_year1),
            'FV1_Deposit': int(FV1_deposit),
            'Deposit_rate2 %' : round(deposit_rate2 * 100, 1),
            'Rate_business2 %': round(b * 100, 1),
            'FV2_Business': int(income_year2),
            'FV2_Deposit': int(FV2_deposit),
            'Deposit_rate3 %' : round(deposit_rate3 * 100, 1),
            'Rate_business3 %': round(c * 100, 1),
            'FV3_Business': int(income_year3),
            'FV3_Deposit': int(FV3_deposit),
            'Deposit_rate4 %' : round(deposit_rate4 * 100, 1),
            'Rate_business4 %': round(d * 100, 1),
            'FV4_Business': int(income_year4),
            'FV4_Deposit': int(FV4_deposit),
            'Total profit_Business %' : int(Total_profit),
            'Total profit_Deposit %' : int(Total_profit_deposit),
            'Total_win' : Total_profit_deposit - Total_profit
        })
            
        df = pd.concat([df, new_row])
        
df.reset_index(drop=True, inplace=True)

df

```

# Анализ

```python
df_deposit = df.drop(columns=['ID of investor', 'Investor', 'Business', 'Rate_business1 %', 'Rate_business2 %', 'Rate_business3 %', 'Rate_business4 %',
 'FV1_Business', 'FV2_Business', 'FV3_Business', 'FV4_Business', 'Total profit_Business %'])
corr_deposit = df_deposit.corr()
corr_deposit.round(2)
```

# Матрица корреляции

```python
plt.figure(figsize = (10,10))
sns.heatmap(corr_deposit, annot = True, cmap = 'plasma',
            cbar_kws = {'label' : 'correlation coefficient'},
            square = True, fmt ='.2f')
plt.title('Correlation plot of data_deposit', color = 'r', fontsize = 16)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
target_variable = 'Total_win'

X = df_deposit.drop(target_variable, axis=1)
y = df_deposit[target_variable]
```

```python
train_size = 0.8
X_train, X_test, y_train, y_test =  train_test_split(X, y, train_size=train_size, random_state=42, shuffle=True)

# Fit a linear regression model to data
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predict on test set
y_pred = lin_reg.predict(X_test)

#add constant to predictor variables
x_train = sm.add_constant(X_train)

#fit linear regression model
model = sm.OLS(y_train, X_train).fit()

#view model summary
print(model.summary())
```
